```python
bearer_token = os.environ.get("BEARER_TOKEN")
```

```python
def bearer_oauth(r):
    """
    Method required by bearer token authentication.
    """

    r.headers["Authorization"] = f"Bearer {bearer_token}"
    r.headers["User-Agent"] = "v2RecentSearchPython"
    return r
```

```python
def connect_to_endpoint(url, params, next_token):
    params['pagination_token'] = next_token
    response = requests.get(url, auth=bearer_oauth, params=params)
    print(response.status_code)
    if response.status_code != 200:
        raise Exception(response.status_code, response.text)
    return response.json()


def main():
    json_response = connect_to_endpoint(search_url, query_params)
```

```python
query_params = {'query': query,
                "tweet.fields": 'attachments,author_id,context_annotations,conversation_id,created_at,entities,geo,id,in_reply_to_user_id,lang,possibly_sensitive,public_metrics,referenced_tweets,reply_settings,source,text,withheld',
                "expansions": 'attachments.media_keys,attachments.poll_ids,author_id,entities.mentions.username,geo.place_id,in_reply_to_user_id,referenced_tweets.id,referenced_tweets.id.author_id',
                "media.fields": 'alt_text,duration_ms,height,media_key,non_public_metrics,organic_metrics,preview_image_url,promoted_metrics,public_metrics,type,url,variants,width',
                "poll.fields": 'duration_minutes,end_datetime,id,options,voting_status',
                "user.fields": 'created_at,description,entities,id,location,name,pinned_tweet_id,profile_image_url,protected,public_metrics,url,username,verified,withheld',
                "place.fields": 'contained_within,country,country_code,full_name,geo,id,name,place_type',
                "next_token": {},
                "max_results":100,
                "start_time": start_time,
                "end_time":end_time,
#                 "since_id":'1559918724097925122',
                'pagination_token':{}
               }
```

```python
json_response = connect_to_endpoint(search_url, query_params, next_token = None)
```

```python
while True:
    if 'next_token' in json_response['meta']:
        next_token = json_response['meta']['next_token']
#         query_params['next_token'] = next_token
        print (next_token)
        json_response = connect_to_endpoint(search_url, query_params, next_token)
        print(json_response['meta']['result_count'])
        df = pd.DataFrame(json_response['data'])
        results1 = results1.append(df)
        time.sleep(1)
    else:
        json_response = connect_to_endpoint(search_url, query_params, next_token = None)
        print(json_response['meta']['result_count'])
        df = pd.DataFrame(json_response['data'])
        results1 = results1.append(df)
        break
```
